```python
def detect_outliers_iqr(data,col):
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1- 1.5*IQR
    upper=Q3+ 1.5*IQR
    outliers=data[(data[col]<lower) | (data[col]>upper)][col]
    return outliers
```

```python
print('Salary outliers (IQR):')
print(detect_outliers_iqr(df, 'Salary'))
```

```python
print('Age outliers (IQR):')
print(detect_outliers_iqr(df, 'Age'))
```

```python
def detect_outliers_zscore(data,col):
    mean=data[col].mean()
    std=data[col].std()
    zscore=(data[col]-mean)/std
    outliers=data[np.abs(zscore)>3][col]
    return outliers
```

```python
print('Salary outliers (zscore):')
print(detect_outliers_zscore(df,'Salary'))
```

```python
print('Age outliers (zscore):')
print(detect_outliers_zscore(df,'Age'))
```

```python
from sklearn.preprocessing import PowerTransformer

df_transformed = df.copy()

for col in numerical_cols:
    if df[col].isnull().sum() < len(df): 
        skew_val = df[col].skew()
        print(f"\nColumn: {col}, Skewness before: {skew_val}")

        if skew_val > 0.5:  # Right skew
            #Log
            if (df[col] > 0).all():
                df_transformed[col+"_log"] = np.log1p(df[col])  
                print(f"Applied Log to {col}")
                
            #Box-Cox
            if (df[col] > 0).all():
                pt = PowerTransformer(method="box-cox")
                df_transformed[col+"_boxcox"] = pt.fit_transform(df[[col]])
                print(f"Applied Box-Cox to {col}")
                
            #Yeo-Johnson
            pt = PowerTransformer(method="yeo-johnson")
            df_transformed[col+"_yeojohnson"] = pt.fit_transform(df[[col]])
            print(f"Applied Yeo-Johnson to {col}")

        elif skew_val < -0.5:  # Left skew
            pt = PowerTransformer(method="yeo-johnson")
            df_transformed[col+"_yeojohnson"] = pt.fit_transform(df[[col]])
            print(f"Applied Yeo-Johnson to {col}")

        else:
            print(f"No transformation needed for {col}")
```
